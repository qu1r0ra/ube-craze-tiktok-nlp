# Ube Craze TikTok Scraping

This notebook demonstrates how to load target video URLs from `links.txt` and execute the Playwright network interception scraper to gather video metadata, comments, and direct replies.

## 1. Setup & Imports

```python
from ube_craze_nlp.scraper.engine import TikTokScraperEngine, read_video_urls
from ube_craze_nlp.utils.paths import LINKS_FILE, RAW_DATA_DIR, ensure_dirs

# Ensure required directories exist
ensure_dirs()
print(f"Raw data will be saved to: {RAW_DATA_DIR}")
```

## 2. Load Target URLs

The target URLs are loaded from `links.txt` at the root of the workspace. Add your 30 selected video URLs there before running.

```python
urls = read_video_urls()
print(f"Loaded {len(urls)} target URLs from {LINKS_FILE.name}.")
for idx, url in enumerate(urls, 1):
    print(f"{idx}. {url}")
```

## 3. Run Scraping Engine

We initialize the scraping engine. Set `headless=False` if you want to watch Playwright scroll and click comment expansions in real-time.

```python
# Initialize scraper. Headless=False lets you observe the browser interaction.
engine = TikTokScraperEngine(headless=False, delay_range=(2.0, 4.0))


async def run_scraper():
    for i, url in enumerate(urls, 1):
        print(f"\n--- Scraping Video {i}/{len(urls)} ---")
        try:
            res = await engine.scrape_video(url)
            print(f"Status: {res['status']}")
            print(f"Author: @{res.get('author')}")
            print(f"Comments intercepted: {res.get('comments_count')}")
            print(f"Video MP4 downloaded: {res.get('video_downloaded')}")

            # Random sleep between videos to prevent IP flagging
            await engine.random_sleep(6.0, 12.0)
        except Exception as e:
            print(f"❌ Error scraping {url}: {e}")


# Run the async scraper loop
await run_scraper()
```
